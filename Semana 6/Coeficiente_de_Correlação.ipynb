# Coeficiente de Correlação

**Dados:** Os dados usados neste algoritmo foram obtidos através do site do [INMET](https://portal.inmet.gov.br/dadoshistoricos), e são referente a cidade de Cuiabá-MT no ano de 2021.

**Objetivo:** Calcular a correlação UMIDADE X TEMPERATURA.

## Formula
$r = \frac{\sum_{i=1}^{n} (x_{i} - \bar{x})(y_{i} - \bar{y})}{\sqrt{\sum_{i=1}^{n} (x_{i} - \bar{x})^{2}} \sqrt{\sum_{i=1}^{n} (y_{i} - \bar{y})^{2}}}$

```python
import numpy as np
import pandas as pd

dados = pd.read_csv("INMET_CUIABA.CSV", sep=";", header=8, decimal=",", 
                    encoding="latin_1")
# Correlação de Pearson
x = dados["TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)"]
y = dados["UMIDADE RELATIVA DO AR, HORARIA (%)"]

xMedia = x.mean()
yMedia = y.mean()

r = np.sum((x - xMedia) * (y - yMedia))/(np.sqrt(np.sum((x-xMedia)**2)) * 
                                         np.sqrt(np.sum((y-yMedia)**2)))
print("Correlação de Pearson Temperatura x umidade:", r)
```

Obs: A correlação de Pearson, é um método destinado a correlações lineares.
Existe também a possibilidade de gerar uma matriz de correlação através do método `.corr()`, observe no exemplo a seguir.

```python
# Podemos observar na matriz que o valor gerado confere com o calculado 
# anteriormente.
dados.corr() 
```

Caso seja necessário, é possível gerar um mapa de calor da tabela acima, veja a seguir.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(30, 30))
sns.heatmap(dados.corr(), annot=True, linewidths=0.01, cmap='coolwarm', cbar=True)
plt.show()
```
