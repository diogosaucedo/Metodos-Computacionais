# Regressão Linear
**Dados:** Os dados usados neste algoritmo foram obtidos através do site do [INMET](https://portal.inmet.gov.br/dadoshistoricos), e são referente a cidade de Cuiabá-MT no ano de 2021.

**Objetivo:** Aplicar Regressão Linear nos campos UMIDADE X TEMPERATURA.

## Formulas
$a = \bar{y} - b\bar{x}$

$b = \frac{n(\sum xy) - \sum x  \sum y}{n(\sum x^{2}) - (\sum x)^{2}}$


```python
import numpy as np
import pandas as pd

dados = pd.read_csv('INMET_CUIABA.CSV', sep=';', header=8, decimal=',', 
                    encoding='latin_1')
# Eliminando campos vazios
sub = dados[['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)', 'UMIDADE RELATIVA DO AR, HORARIA (%)']].dropna()

x = sub['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)']
y = sub['UMIDADE RELATIVA DO AR, HORARIA (%)']
n = x.size

# Aplicando na formula
b = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2)-np.sum(x)**2)
a = y.mean() - b * x.mean()

# Fazendo predição
sub['previsto'] = a + b * x

sub['UMIDADE RELATIVA DO AR, HORARIA (%)'].plot(figsize=(16,9))
sub['previsto'].plot()
```
